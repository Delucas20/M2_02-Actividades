**1) Realiza un programa que lea tres números por teclado y permita evaluar lo siguiente:** 
* Si los números están en orden ascendente
* Si los números no están en orden ascendente
* Si hay un error debido a que el primer número introducido es 0

__No uses listas, eso dejalo para el siguiente ejercicio__

```python
valor1 = int(input("Introduce un valor: "))
valor2 = int(input("Introduce otro valor: "))
valor3 =  int(input("Introduce un último valor: "))
print (valor1, valor2, valor3)
if valor1<valor2<valor3:
    print("Los valores están ordenados de forma ascendente")
else:
    print("Los valores no están ordenados de forma ascendente")
if valor1==0: 
    print("El primer valor es 0, y no hay error")
```

**3) Realiza un programa que lea letras y cuente con una variable contador las letras "a" que se introducen. Para salir del programa, introducir el carácter ".". Al finalizar mostrar el número de veces que se ha pulsado la letra "a".**

```python
palabra1 = input("Introduce una palabra: ")
palabra2 = input("Introduce otra palabra: ")
palabra3 = input("Introduce una última palabra: ")
lista = [palabra1, palabra2, palabra3]
print (lista)
contador = 0
for palabra in (palabra1, palabra2, palabra3):
    if 'a' in palabra:
        contador += palabra.count('a')

print("Hay", contador, "letras 'a' en las palabras")
#Como hemos usado "for", no hace falta un programa para salir del bucle


```

**4) Crea una lista de palabras, recorre la lista y muestra cada palabra junto con su longitud. Al final, indicar cual fue la palabra con más caracteres.**

```python

palabra1 = input("Introduce una palabra: ")
palabra2 = input("Introduce otra palabra: ")
palabra3 = input("Introduce una tercera palabra: ")
palabra4 = input("Introduce una cuarta palabra: ")
lista = [palabra1, palabra2, palabra3, palabra4]
for p in lista:
    print(p, '->', len(p))
mas_larga = max(lista, key=len)
print('La palabra con más caracteres es:', mas_larga, 'con', len(mas_larga), 'caracteres')
```
